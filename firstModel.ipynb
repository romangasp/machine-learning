```python
import pandas as pd
data = pd.read_csv("./sample_data/celsius.txt")
```

```python
data.info();
```

```python
data.head()
```

```python
import seaborn as sb

sb.scatterplot(x="celsius", y="fahrenheit", data=data, hue="fahrenheit", palette="coolwarm")
```

```python
#Characteristichs (X), etiqueta(y)


X= data["celsius"]
y= data["fahrenheit"]
```

```python
X.values
```

```python
X_processed = X.values.reshape(-1, 1)
y_processed = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_processed,y_processed)
```

```python
celsius = 7900
prediction = model.predict([[celsius]])
print(f"{celsius} grados celsius son {prediction} grados fahrenheit")
```

```python
model.score(X_processed,y_processed)
```
