```python
import pandas as pd
data = pd.read_csv("./weigth.txt")
```

```python
data.info()
```

```python
data.head()
```

```python
import seaborn as sb
sb.scatterplot(x="altura", y="peso",data=data, hue="peso", palette="coolwarm")
```

```python
X= data["altura"]
y= data["peso"]
```

```python
X.values
```

```python
X_processed = X.values.reshape(-1,1)
y_processed = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_processed,y_processed)
```

```python
altura = 1.85
prediction = model.predict([[altura]])
print(f"{altura} metros son {prediction} kilogramos")
```

```python
model.score(X_processed,y_processed)
```
